# Import Libraries

```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

# Read Dataset File Containing The Housing Information

```python
KaggleHousingDataset = pd.read_csv("Housing.csv")
KaggleHousingDataset
```

# Distribute The Dataset into Two Seperate Datasets

Y includes the SalePrice of the Houses, X includes the area, number of bedrooms, and number of houses

```python
Y = KaggleHousingDataset.SalePrice
Y
```

```python
X = KaggleHousingDataset.drop(['SalePrice'], axis=1)
X
```

# Split Data into Training and Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

# Run the Linear Regression Model

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

# Predict the Prices using the Test Dataset

```python
Y_prediction = model.predict(X_test)
```

```python
print('Coefficients: ', model.coef_)
print('Intercept: ', model.intercept_)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.regplot(x=Y_test, y=Y_prediction)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.show()
```
